#Python questions


1. What is Python, and why is it popular?

Python is a high-level, interpreted programming language known for its simple syntax and readability.
It is popular because it is:

Easy to learn and use

Versatile (used in web development, data science, AI, automation, etc.)

Supported by a large community and rich libraries

2. What is an interpreter in Python?

An interpreter executes Python code line by line, converting it into machine-readable instructions at runtime. This makes debugging easier compared to compiled languages.

3. What are pre-defined keywords in Python?

Keywords are reserved words with special meanings in Python.
Examples: if, else, for, while, def, class, True, False, None.

4. Can keywords be used as variable names?

 No. Keywords cannot be used as variable names because they are reserved for Python’s syntax.

5. What is mutability in Python?

Mutability refers to whether an object’s value can be changed after creation.

Mutable: list, dict, set

Immutable: int, float, string, tuple

6. Why are lists mutable, but tuples are immutable?

Lists are designed to be modifiable (add, remove, change elements), while tuples are fixed to ensure data integrity and allow faster access.

7. Difference between == and is operators?

== → checks value equality

is → checks object identity (same memory location)

Example:

a = [1, 2]
b = [1, 2]
a == b   # True
a is b   # False

8. What are logical operators in Python?

Logical operators are used to combine conditions:

and

or

not

9. What is type casting in Python?

Type casting is the process of converting one data type into another.

Example:

int("10")
float(5)
str(100)

10. Difference between implicit and explicit type casting?

Implicit: Python automatically converts data types
Example: 3 + 2.5 → 5.5

Explicit: Programmer manually converts data types
Example: int("5")

11. Purpose of conditional statements in Python?

Conditional statements allow a program to make decisions based on conditions.

Example:

if age >= 18:
    print("Adult")

12. How does the elif statement work?

elif allows checking multiple conditions after an if.

Example:

if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")

13. Difference between for and while loops?

for loop → used when the number of iterations is known

while loop → used when iterations depend on a condition

14. Scenario where a while loop is more suitable than a for loop

A while loop is better when the end condition is unknown, such as:

Reading input until the user types “exit”

Waiting for a condition to become true

```python
def modify_list(lst):
    lst.append(4)

def modify_number(num):
    num += 1

my_list = [1, 2, 3]
my_num = 10

modify_list(my_list)
modify_number(my_num)

print("List after function:", my_list)
print("Number after function:", my_num)
```

```python
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers:", total)
```

```python
text = input("Enter a string: ")
reversed_text = ""
i = len(text) - 1

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
num = int(input("Enter a number: "))
factorial = 1

while num > 0:
    factorial *= num
    num -= 1

print("Factorial:", factorial)
```
